```python
import plotly
import pandas as pd
# Load data
#df = pd.read_csv("finance-charts-apple.csv")
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/Move%20to%20Canada.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]
print(df)

import plotly.graph_objects as go
line1 = go.Scatter(x=df['Date'], y=df['From US'], name='Close')
fig  = go.Figure(line1)
fig.show()
plotly.offline.plot(fig, filename='lifeExp.html')

#x=plotly.offline.plot(fig, include_plotlyjs=False, output_type='div')
#text_file = open("sample.txt", "w")
#n = text_file.write(x)
#text_file.close()
```
